```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('automobile.csv')

data.head()
```

```python
ds = data.copy()
```

```python
ds.isnull().sum()
```

```python
ds.horsepower.sample(200).values
```

```python
cat = data.select_dtypes(exclude = "number")
num = data.select_dtypes(include = 'number')

print('Numerical Data')
display(num.head())
print(f'\n\nCategorical Data')
display(cat.head())
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
for i in data.columns:
    if ((data[i].isnull().sum() / len(data)) * 100) < 30:
        if data[i].dtypes != 'O':
            data[i].fillna(data[i].median(), inplace = True)
        else:
            data[i].fillna(data[i].mode()[0], inplace = True)
    else:
        data.drop(i, axis = 1, inplace = True)

data.isnull().sum()
```

```python
cols_to_scale = []
for i in num.columns:
    if data[i].describe()[2] > 1500:
        cols_to_scale.append(i)

cols_to_scale
```

```python
# LABEL ENCODE THE CATEGORICAL COLUMNS
from sklearn.preprocessing import LabelEncoder
import joblib

encoded = {}  # ................................................................ Have a dictionary to save the encoders
for i in cat.columns: # ........................................................ Iterate through the categorical variables
    encode = LabelEncoder() # .................................................. Instantiate the encoder transformer
    data[i] = encode.fit_transform(data[i]) # ...................................... Encode the columns in ech iteration
    joblib.dump(encode, open(f'{i}_encoder.pkl', 'wb')) # ...................... Save the transformer to file for future reusal
    encoded[i +'_encoder'] = encode # .......................................... Save the transformer to the above stated dictionary for easy access

encoded
```
